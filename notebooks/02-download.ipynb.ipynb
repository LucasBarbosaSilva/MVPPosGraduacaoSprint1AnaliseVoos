```python
import os
import subprocess

meses = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
anos = {
    "2020": meses,
    "2021": meses,
    "2022": meses,
    "2023": meses,
    "2024": meses,
}

destino_base = "/Volumes/mvp/staging/anac/zip"

os.makedirs(destino_base, exist_ok=True)

for ano, meses in anos.items():
    for mes in meses:
        nome_arquivo = f"basica{ano}-{mes}.zip"
        url = f"https://www.gov.br/anac/pt-br/assuntos/regulados/empresas-aereas/Instrucoes-para-a-elaboracao-e-apresentacao-das-demonstracoes-contabeis/envio-de-informacoes/basica/{ano}/{nome_arquivo}"
        destino = os.path.join(destino_base, nome_arquivo)

        print(f"Baixando {url} -> {destino}")
        subprocess.run([
            "wget",
            "--user-agent=Mozilla/5.0",
            "-O", destino,
            url
        ])
```
