## Quantidade de voos realizado por empresa aérea, por rota e variando por mês, semestre e ano:

```python
%sql
select id_empresa, nm_empresa from dim_empresa;
```

```python
%sql
select sg_iata, nm_aerodromo, sg_uf, nm_pais, nm_municipio from mvp.gold.dim_aeroporto;
```

```python
from pyspark.sql.window import Window
from pyspark.sql import functions as F

id_empresa = input("Digite o id da empresa ('' para todos):")
sg_aeroporto_origem = input("Digite a sigla do aeroporto de origem ('' para todos):")
sg_aeroporto_destino = input("Digite a sigla do aeroporto de destino ('' para todos):")
ano = input("Digite o ano (2020-2024) ('' para todos):")
semestre = input("Digite o semestre (1-2) ('' para todos):")
mes = input("Digite o mês (1-12) ('' para todos):")

df = spark.sql(f"""
                select 
                    ap.sg_iata AS origem, 
                    ac.sg_iata as destino, 
                    d.nr_ano as ano, 
                    d.nr_semestre as semestre, 
                    d.nr_mes as mes, 
                    count(*) as qtde_voos
                from fact_voo v 
                inner join dim_aeroporto ap on v.pk_id_aerodromo_partida = ap.id_aerodromo
                inner join dim_aeroporto ac on v.pk_id_aerodromo_chegada = ac.id_aerodromo
                inner join dim_data d on v.pk_id_dt_partida = d.dt_voo 
                where v.pk_id_empresa LIKE '%{id_empresa}%' 
                AND ap.sg_iata LIKE '%{sg_aeroporto_origem}%'
                AND ac.sg_iata LIKE '%{sg_aeroporto_destino}%'
                AND d.nr_mes LIKE '%{mes}%'
                AND d.nr_semestre LIKE '%{semestre}%'
                AND d.nr_ano LIKE '%{ano}%'
                group by origem, destino, mes, ano, semestre
                order by ano, semestre, ano, semestre, mes
            """)
print("Total de voos encontrados para o período selecionado:", df.select(F.sum("qtde_voos").alias("total")).collect()[0].total)
df.display()

```

## Total de passageiros transportados por empresa aérea, por rota e variando por mês, semestre e ano;

```python
from pyspark.sql.window import Window
from pyspark.sql import functions as F

id_empresa = input("Digite o id da empresa ('' para todos):")
sg_aeroporto_origem = input("Digite a sigla do aeroporto de origem ('' para todos):")
sg_aeroporto_destino = input("Digite a sigla do aeroporto de destino ('' para todos):")
ano = input("Digite o ano (2020-2024) ('' para todos):")
semestre = input("Digite o semestre (1-2) ('' para todos):")
mes = input("Digite o mês (1-12) ('' para todos):")

nm_empresa = spark.sql(f"""
                select 
                    nm_empresa
                from dim_empresa
                where id_empresa LIKE '%{id_empresa}%'
            """).collect()[0].nm_empresa
print("Buscando resultados para a empresa:", nm_empresa)

df = spark.sql(f"""
                select 
                    ap.sg_iata AS origem, 
                    ac.sg_iata as destino, 
                    d.nr_ano as ano, 
                    d.nr_semestre as semestre, 
                    d.nr_mes as mes, 
                    sum(v.nr_passag_pagos) + sum(v.nr_passag_gratis) as tt_passageiros
                from fact_voo v 
                inner join dim_aeroporto ap on v.pk_id_aerodromo_partida = ap.id_aerodromo
                inner join dim_aeroporto ac on v.pk_id_aerodromo_chegada = ac.id_aerodromo
                inner join dim_data d on v.pk_id_dt_partida = d.dt_voo 
                where v.pk_id_empresa LIKE '%{id_empresa}%' 
                AND ap.sg_iata LIKE '%{sg_aeroporto_origem}%'
                AND ac.sg_iata LIKE '%{sg_aeroporto_destino}%'
                AND d.nr_mes LIKE '%{mes}%'
                AND d.nr_semestre LIKE '%{semestre}%'
                AND d.nr_ano LIKE '%{ano}%'
                group by origem, destino, mes, ano, semestre
                order by ano, semestre, ano, semestre, mes
            """)
print("Total de passageiros:", df.select(F.sum("tt_passageiros").alias("total")).collect()[0].total)
df.display()

```

## Quais as rotas com maior custo benefício baseado na relação entre passageiros pagantes e consumo de combustível, variando por mês, semestre e ano;

```python
ano = input("Digite o ano (2020-2024) ('' para todos):")
semestre = input("Digite o semestre (1-2) ('' para todos):")
mes = input("Digite o mês (1-12) ('' para todos):")

df = spark.sql(f"""
        select 
            ap.sg_iata AS origem,
            ap.nm_aerodromo as aeroporto_origem,
            ap.sg_uf as uf_origem,
            ac.sg_iata as destino, 
            ac.nm_aerodromo as aeroporto_destino,
            ac.sg_uf as uf_destino,
            d.nr_ano as ano, 
            d.nr_semestre as semestre,
            d.nr_mes as mes,
            sum(v.lt_combustivel) litros_combustivel, 
            sum(v.nr_passag_pagos) passageiros_pagos, 
            (sum(v.lt_combustivel)/sum(v.nr_passag_pagos)) as custo_beneficio
        from fact_voo v
        inner join dim_aeroporto ap on v.pk_id_aerodromo_partida = ap.id_aerodromo
        inner join dim_aeroporto ac on v.pk_id_aerodromo_chegada = ac.id_aerodromo
        inner join dim_data d on v.pk_id_dt_partida = d.dt_voo
        WHERE d.nr_mes LIKE '%{mes}%'
        AND d.nr_semestre LIKE '%{semestre}%'
        AND d.nr_ano LIKE '%{ano}%'
        group by origem, aeroporto_origem, uf_origem, destino, aeroporto_destino, uf_destino, ano, semestre, mes
        order by custo_beneficio
        limit 10;
    """)

df.display()
```
